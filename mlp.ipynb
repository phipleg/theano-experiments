# Multi layer perceptron

```python
import numpy as np
from sklearn import datasets
import theano
import theano.tensor as T
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

# Use float32 as the default float data type
theano.config.floatX = 'float32'
# Display plots inline and change default figure size
%matplotlib inline
sns.set(color_codes=True)
np.random.seed(0)
```

```python
n_samples = 500
n_features = 2
n_hidden = 20
n_classes = 3
train_X, train_y = datasets.make_classification(n_samples=n_samples,
                                                n_features=n_features,
                                                n_redundant=0,
                                                n_informative=2,                             
                                                n_clusters_per_class=1, 
                                                n_classes=n_classes)
train_y_onehot = np.eye(n_classes)[train_y]

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Generated items')
ax.scatter(train_X[:, 0], train_X[:, 1], marker='o', c=train_y)
plt.show()
```

```python
# Define model
X = theano.shared(train_X.astype('float32'))
y = theano.shared(train_y_onehot.astype('float32'))

W1 = theano.shared(np.random.randn(n_features, n_hidden).astype('float32'), name='W1')
b1 = theano.shared(np.zeros(n_hidden).astype('float32'), name='b1')
W2 = theano.shared(np.random.randn(n_hidden, n_classes).astype('float32'), name='W1')
b2 = theano.shared(np.zeros(n_classes).astype('float32'), name='b1')

z1 = X.dot(W1) + b1
a1 = T.tanh(z1)
z2 = a1.dot(W2) + b2
y_hat = T.nnet.softmax(z2)
prediction = T.argmax(y_hat, axis=1)

reg_lambda = np.float32(0.01)
loss1 = T.nnet.categorical_crossentropy(y_hat, y).mean() # use cross-entropy loss for softmax classification
loss2 = reg_lambda/2.0 * (T.sum(T.sqr(W1)).mean() + T.sum(T.sqr(W2)).mean())  # add L2-regularization
loss = loss1 + loss2
```

```python
# Gradient descent
dW1 = T.grad(loss, W1)
db1 = T.grad(loss, b1)
dW2 = T.grad(loss, W2)
db2 = T.grad(loss, b2)

epsilon = np.float32(0.01)
gradient_step = theano.function([],
                                updates=((W1, W1 - epsilon * dW1),
                                         (b1, b1 - epsilon * db1),
                                         (W2, W2 - epsilon * dW2),
                                         (b2, b2 - epsilon * db2)))

# Define loss function in order to access shared variable
f_loss = theano.function([], loss)
```

```python
# build model
n_epochs = 1000
for i in xrange(n_epochs):
    gradient_step()
    if i % 100 == 0:
        print i, 'loss=',f_loss()
```

```python
f_prediction = theano.function([], prediction, allow_input_downcast=True)
```

```python
y_pred = f_prediction()
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Predictions')
ax.scatter(train_X[:, 0], train_X[:, 1], marker='o', c=y_pred)
plt.show()
```

```python
error = np.mean(np.abs(np.sign(train_y-y_pred)))
```
